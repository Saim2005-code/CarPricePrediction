```python
df['transmission']=df['transmission'].map({'Automatic':1,'Manual':0})
df['fuelType']=df['fuelType'].map({'Petrol':1,'Diesel':0})
df
```

```python
df.info()
```

```python
df['model']=df['model'].str.strip()
```

```python
models=list(df['model'])
models
```

```python
new_df=pd.get_dummies(df,columns=['model'])
```

```python
cleaned_df=new_df.replace(True,1)
cleaned_df=cleaned_df.replace(False,0)
cleaned_df
```

```python
cleaned_df.describe()
```

```python
cleaned_df.isna().sum()
```

```python
trans_mean=df['transmission'].mean(numeric_only=True)
trans_mean
```

```python
cleaned_df['transmission']=cleaned_df['transmission'].fillna(trans_mean)
cleaned_df.isna().sum()
```

```python
fuelType_mean=cleaned_df['fuelType'].mean(numeric_only=True)
cleaned_df['fuelType']=cleaned_df['fuelType'].fillna(fuelType_mean)
cleaned_df.isna().sum()
```

```python
cleaned_df
```

```python
plot_list=['price','transmission','fuelType']
new_plot_list=['mileage','tax','mpg','engineSize']
```

```python
def plotting(plot_list):
    for i in range(0,len(plot_list)):
        plt.subplot(2,2,i+1)
        sns.histplot(cleaned_df[plot_list[i]],bins=20,kde=True)
```

```python
plotting(plot_list)
plt.tight_layout()
```

```python
plotting(new_plot_list)
plt.tight_layout()
```

```python
def normalization(x,s):
    m=cleaned_df['mileage'].mean(numeric_only=True)
    d=cleaned_df['mileage'].std(numeric_only=True)
    result=(x-m)/d
    return result
```

```python
cleaned_df['mileage']=cleaned_df['mileage'].apply(normalization).round(2)
cleaned_df
```

```python
sns.histplot(cleaned_df['mileage'],bins=100,kde=True)
plt.plot()
```

```python
cleaned_df.head()
```

```python
mpg_list=list(cleaned_df['mpg'])
def new_normal(x):
    m=cleaned_df['mpg'].mean(numeric_only=True)
    d=cleaned_df['mpg'].std(numeric_only=True)
    result=(x-m)/d
    return result
new_mpg_list=cleaned_df['mpg'].apply(new_normal).round(2)
new_mpg_list
```

```python
sns.histplot(new_mpg_list,bins=30,kde=True)
plt.plot()
```

```python
cleaned_df['mpg']=new_mpg_list
cleaned_df.head()
```

```python
from sklearn.model_selection import train_test_split
y=cleaned_df['price']
x=cleaned_df.drop(['price'],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_predict)
print(f'The R² accuracy is: {r2*100//1}%')
```

```python
n=x_test.shape[0]
k=x_test.shape[1]
adj_r2=1-((1-r2)*(n-1)/(n-k-1))
print(f'The Adjusted R² accuracy is: {adj_r2*100//1}%')
```
